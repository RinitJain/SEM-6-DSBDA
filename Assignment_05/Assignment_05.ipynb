```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('titanic.csv')
data
```

```python
data['Age'].describe()
```

```python
data['Sex'].unique()
```

```python
data.isnull().any()
```

```python
data_clean = data.dropna(subset=['Age'])
data_clean
```

```python
#Running boxplot on Original data

plt.figure(figsize=(8,6))
sns.boxplot(data=data, x='Sex', y='Age', hue='Survived')
plt.title("Age Distribution by Gender and Survival Status on Titanic")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

```python
#Running Boxplot on Cleaned data (by age)

plt.figure(figsize=(8,6))
sns.boxplot(data=data_clean, x='Sex', y='Age', hue='Survived')
plt.title("Age Distribution by Gender and Survival Status on Titanic")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```
