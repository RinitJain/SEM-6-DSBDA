```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("nba.csv")
data
```

```python
data.dtypes
```

```python
data.isnull().any().sum()
```

```python
#Converts text-based value into a Category(i.e.datatype) to save memory
#Internally, it stores the data as integer codes and a lookup table of the actual category names.

data["Position"] = data["Position"].astype("category")
```

```python
def height_toInches(height):
  if pd.isnull(height) : return np.nan
  feet, inches = height.split("-")
  return int(feet)*12 + int(inches)

data["Height"] = data["Height"].apply(height_toInches)
```

```python
data["Salary"] = data["Salary"].fillna(data["Salary"].median())
data["Number"] = data["Number"].fillna(data["Number"].median())
data["Height"] = data["Height"].fillna(data["Height"].median())
data["Weight"] = data["Weight"].fillna(data["Weight"].median())
```
