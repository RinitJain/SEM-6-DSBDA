```python
# Provide summary statistics (mean, median, minimum, maximum, standard
# deviation) for a dataset (age, income etc.) with numeric variables grouped by
# one of the qualitative (categorical) variable. For example, if your categorical
# variable is age groups and quantitative variable is income, then provide
# summary statistics of income grouped by the age groups. Create a list that
# contains a numeric value for each response to the categorical variable.
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('nba.csv')
data
```

```python
data.describe()
```

```python
data_clean = data.dropna(subset=['Team','Salary'])
data_clean
```

```python
# 1. Group the data
grouped_data = data_clean.groupby('Team')['Salary']

# 2. Aggregate 
summary = grouped_data.agg(['mean', 'median', 'min', 'max', 'std'])
```

```python
# 3. Reset index
summary_stats = summary.reset_index()
summary_stats
```

```python
#Now lets group by the age 

bins = [19, 25, 30, 35, 40]
labels = ['19-24', '25-29', '30-34', '35-39']

age_groups = pd.cut(data['Age'], bins=bins, labels=labels, right=True)

age_groups.value_counts()
```

```python
data_clean['Age Groups'] = age_groups
data_clean
```

```python
stats_by_age_group = data_clean.groupby('Age Groups')['Salary']

grouped_stats= stats_by_age_group.agg(['mean', 'median', 'std'])

stats = grouped_stats.reset_index()

stats
```
